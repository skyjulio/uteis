```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario_marcos = np.array([150,210,377,400])
salario_andre = np.array([50,300,510,450])
```

```python
plt.plot(salario_marcos)
plt.plot(salario_andre, ls='--', ms = 8, marker='^', label = 'salario andre')
plt.plot(salario_andre, c='red', marker='s')
plt.legend() #loc='upper left'
plt.show()
```
